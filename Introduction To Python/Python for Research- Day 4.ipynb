# Calculate the mean of the follwing list

````python
scores = [20, 22, 59, 49, 27, 0, 62, 40, 58, 92, 36,  63, 75, 50, 73, 7, 49, 42, 1, 79]
````

````python
total = 0

for item in scores:
    total = total + item

mean = total / len(scores)
print("Mean = {}".format(mean))
````

# List Comprehension

````python
nums = [3, 4, 5, 11, 6]

sq = []
for item in nums:
    sq.append(item ** 2)
````

# Syntax for List comprehension
## [expression for item in iterables]

````python
[item ** 2 for item in nums]
````
