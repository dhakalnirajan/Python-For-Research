# Task

```python
# WAP to take input from the user and check if the input number is positive, negative or zero.
a = int(input("Enter a number : "))
if a > 0:
    print("The Number is positive.")
elif a < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

```python
# Nesting of Conditionals
user_input = int(input("Enter a number : "))
if user_input == 0:
    print("zero")
elif user_input < 0:
    print("negative")
else:
    print("positive")
```
