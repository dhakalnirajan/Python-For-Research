# Line Plot

```python
x = np.linspace(0, 15, 30)
y = np.sin(x) + 0.1 * np.random.randn((len(x)))
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, lw = 2, color = 'purple') # lw = line-width
```

```python
plt.plot(x,y,'--' , lw = 2, color = 'blue') # lw = line-width
```

```python
plt.plot(x,y,'o--' , lw = 2, color = 'purple') # lw = line-width
```

```python
plt.plot(x,y,'x--' , lw = 2, color = 'purple', ms = 8) # ms = marker strength
```

```python
# change figure size
plt.figure(figsize = (5,6))
plt.plot(x,y,'--' , lw = 2, color = 'purple') # lw = line-width
```

```python
# Label axis
plt.figure(figsize = (8,8))
plt.plot(x,y,'o--' , lw = 2, color = 'purple', ms= 8)   # lw - lengthwidth
plt.xlabel('Time [s]', fontsize = 20)
plt.ylabel('Voltage[v]')
plt.title("Voltage in Electrode")
plt.show()
```

```python
# Adding Legend to graph
plt.figure(figsize = (8,8))
plt.plot(x,y,'o--' , lw = 2, color = 'purple', ms= 8, label = 'Comp 1')   # lw - lengthwidth
plt.xlabel('Time [s]', fontsize = 20)
plt.ylabel('Voltage[v]')
plt.title("Voltage in Electrode")
plt.legend(loc= 'lower right', fontsize = 13)
plt.show()
```

```python
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,'o', label = 'Data')
plt.plot(x2, y2, label = "Fit")
plt.xlabel('Time [s]', fontsize = 20)
plt.ylabel('Voltage[v]')
plt.title("Voltage in Electrode")
plt.legend(loc= 'lower right', fontsize = 13)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,'o', label = 'Data')
plt.plot(x2, y2, label = "Fit")
plt.xlabel('Time [s]', fontsize = 20)
plt.ylabel('Voltage[v]')
plt.title("Voltage in Electrode")
plt.legend(loc= 'lower right', fontsize = 13)
plt.xlim(left= -1 ,right = 10) # limits curve upto 10 on x axis 
plt.ylim(top = 2) # limits curve upto 2 on y axis
plt.show()
```

# Histogram

```python
res = np.random.randn(1000) * 0.2 + 0.4
```

```python
# Plotting Histogram

plt.hist(res)
```

```python
plt.figure(figsize= (8,3))
plt.hist(res, bins = 'auto')
```

```python
plt.figure(figsize= (8,3))
plt.hist(res, bins = 30) # bin means interval
plt.show()
```

```python
# This is example of another plotting library
import plotly.express as px
```

```python
px.histogram(res)
```

# Density Plot : Histogram plot when integrated area under histogram 1

```python
plt.figure(figsize= (5,3))
plt.hist(res, bins = 30, density = True)
plt.show()
```

# Plot two histogram in one

```python
res2 = np.random.randn(100) * 0.2 + 0.4
```

$\hat{E} - e $

```python
plt.figure(figsize= (8,3))
plt.hist(res, bins = 30, density = True, histtype = 'step', label = "Entropy")
plt.hist(res2, bins = 30, density = True, histtype = 'step', label = 'Frequency')
plt.xlabel("$\hat{E}- E$", fontsize = 20, ) # $ "Latex " $
plt.ylabel('Frequency')
plt.legend(loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize= (8,3))
plt.hist(res, bins = 30, density = True, histtype = 'step', label = "Entropy")
plt.hist(res2, bins = 30, density = True, histtype = 'step', label = 'Frequency')
plt.xlabel("$\hat{E}- E$", fontsize = 20, ) # $ "Latex " $
plt.ylabel('Frequency')
plt.legend(loc = 'upper right')
plt.savefig('trail.png', dpi = 400)

plt.show()
```

# Adding text in image

```python
np.random.seed(0)
res1 = np.random.randn(100) * 0.3 + 0.2
```

```python
plt.figure(figsize= (8,3))
plt.hist(res, bins = 30, density = True, histtype = 'step', label = 'Method 1')
plt.legend(loc = 'upper right')
plt.show()
```

```python
std = round(np.std(res1),3)
std
```

```python
plt.figure(figsize= (8,3))
plt.hist(res, bins = 30, density = True, histtype = 'step', label = 'Method 1')
plt.legend(loc = 'upper right')
plt.text(0,1.5,f"$\sigma_a = ${std}", bbox = dict(facecolor= 'white', edgecolor = 'black'))   # $ "Latex " $
plt.show()
```
