```python
x = np.linspace(0, 10, 1000)

sin = np.sin(x)

lin = np.arange(1, 1000, 10)

cos = np.cos(x)

tan = np.tan(x)
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,8))
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,8))

# Sine Wave
ax = axes[0,0]
ax.plot(x, sin, color = 'r')

# Regression Fit
ax = axes[0,1]
ax.plot(lin, marker = 'o', color = 'b')

# Tangent Wave
ax = axes[1,0]
ax.plot(x, tan, color = 'g')

# Cosine Wave
ax = axes[1,1]
ax.plot(x, cos, color = 'purple')
```

```python
# Adding Legends
fig, axes = plt.subplots(2,2, figsize=(10,8))

# Sine Wave
ax = axes[0,0]
ax.plot(x, sin, color = 'r', label = 'Sinx')
ax.legend()
ax.set_title("Sinx Function")

# Regression Fit
ax = axes[0,1]
ax.plot(lin, marker = 'o', color = 'b', label = 'Reg')
ax.legend()
ax.set_title("Regression Fit")

# Tangent Wave
ax = axes[1,0]
ax.plot(x, tan, color = 'g', label = 'Tanx')
ax.legend()
ax.set_title("Tanx Function")

# Cosine Wave
ax = axes[1,1]
ax.plot(x, cos, color = 'purple', label = 'Cosx')
ax.legend()
ax.set_title("Cosx Function")
```

# Tight Layouting the Images / Padding the Image.

```python
fig, axes = plt.subplots(2,2, figsize=(10,8))

# Sine Wave
ax = axes[0,0]
ax.plot(x, sin, color = 'r', label = 'Sinx')
ax.legend()
ax.set_title("Sinx Function")

# Regression Fit
ax = axes[0,1]
ax.plot(lin, marker = 'o', color = 'b', label = 'Reg')
ax.legend()
ax.set_title("Regression Fit")

# Tangent Wave
ax = axes[1,0]
ax.plot(x, tan, color = 'g', label = 'Tanx')
ax.legend()
ax.set_title("Tanx Function")

# Cosine Wave
ax = axes[1,1]
ax.plot(x, cos, color = 'purple', label = 'Cosx')
ax.legend()
ax.set_title("Cosx Function")

fig.tight_layout(pad=6)
fig.show()
```

```python
# X-label and Y-label

fig, axes = plt.subplots(2,2, figsize=(10,8))

# Sine Wave
ax = axes[0,0]
ax.plot(x, sin, color = 'r', label = 'Sinx')
ax.legend()
ax.set_title("Sinx Function")
ax.set_xlabel("$x$")
ax.set_ylabel("$Sinx$")

# Regression Fit
ax = axes[0,1]
ax.plot(lin, marker = 'o', color = 'b', label = 'Reg')
ax.legend()
ax.set_title("Regression Fit")

# Tangent Wave
ax = axes[1,0]
ax.plot(x, tan, color = 'g', label = 'Tanx')
ax.legend()
ax.set_title("Tanx Function")
ax.set_xlabel("$x$")
ax.set_ylabel("$Tanx$")

# Cosine Wave
ax = axes[1,1]
ax.plot(x, cos, color = 'purple', label = 'Cosx')
ax.legend()
ax.set_title("Cosx Function")
ax.set_xlabel("$x$")
ax.set_ylabel("$Cosx$")

fig.tight_layout(pad=1)
fig.show()
```

# Contour Plot

```python
_ = np.linspace(-1, 1, 100)
x, y = np.meshgrid(_,_)
z = x ** 2 + x * y
plt.contourf(x,y,z)
```

```python
plt.contour(x, y, z, levels = 20)
plt.colorbar(label='Temperature [$^\circ C$')
plt.xlabel("Horizontal Position")
plt.ylabel("Vertical Position")
```

```python
plt.figure(figsize=(10,6))
cs = plt.contour(x,y,z, levels = 20)
plt.clabel(cs, fontsize = 8)
```
