# Create create the series from numpy array

```python
import numpy as np
```

```python
x = np.array(['a', 'e', 'i', 'o', 'u'])
```

```python
s = pd.Series(x)
```

```python
s
```

# Create pandas series with specified index

```python
pd.Series(['bmw', 'audi', 'mercedes'])
```

```python
pd.Series(['bmw', 'audi', 'mercedes'], index = ['first car', 'second car', 'third car'])
```

```python
s = pd.Series(['bmw', 'audi', 'mercedes'], index = ['first car', 'second car', 'third car'])
```

```python
s
```

```python
s.index
```

```python
list(s.index)
```

```python
s.values
```

# Create pandas series using dictionary

```python
d = {
    'a': 40,
    'b': 88,
    'c': 0
}
```

```python
pd.Series(d)
```

# Accessing data from pandas series with position index

```python
s = pd.Series(['bmw', 'audi', 'mercedes'], index = ['first_car', 'second_car', 'third_car'])
```

```python
s
```

```python
s[0]
```

```python
s[1]
```

```python
s['first_car']
```

# giving name to pandas series

```python
s = pd.Series(['bmw', 'audi', 'mercedes'], index = ['first_car', 'second_car', 'third_car'], name = 'cars')
```

```python
s
```
