```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'city': ['New York', 'Chicago', 'Orlando', 'New Jersey'],
    'temperature':[21, 14, 35, 22],
})


df2 = pd.DataFrame({
    'city': ['New York', 'Chicago', 'Orlando'],
    'humidity':[65, 85, 75],
})
```

```python
df1
```

```python
df2
```

```python
df3 = pd.merge(df1, df2, on = 'city')   # merge or join
```

```python
df3
```

# Types of database join

<img src="https://www.datasciencemadesimple.com/wp-content/uploads/2017/09/join-or-merge-in-python-pandas-1.png" align=center>

```python
df3 = pd.merge(df1, df2, on = 'city', how = 'outer')   # default inner join how = 'inner' 
```

```python
df3
```

```python
df3 = pd.merge(df1, df2, on = 'city', how = 'left')   # default inner join how = 'inner' 
```

```python
df3
```

```python
df3 = pd.merge(df1, df2, on = 'city', how = 'right')   # default inner join how = 'inner' 
```

```python
df3
```
