```python
import pandas as pd
```

```python
url = 'http://bit.ly/drinksbycountry'
```

```python
df = pd.read_csv(url)
```

```python
# calculate the mean beer servings for each continent

df.groupby('continent').beer_servings.mean()
```

```python
# calculate mean beer servings for Africa

df.groupby('continent').beer_servings.mean()['Africa']
```

```python
# calculate mean beer servings for Asia

df.groupby('continent').beer_servings.mean()['Asia']
```

```python
# calculate max beer serving continent wise

df.groupby('continent').beer_servings.max()
```

```python
import matplotlib.pyplot as plt

plt.bar(x = df.groupby('continent').beer_servings.max(), height = df.groupby('continent').beer_servings.max(), width= 8)
```

```python
df.groupby('continent').beer_servings.agg(['min', 'max', 'mean', 'count'])
```
