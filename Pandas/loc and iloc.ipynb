# loc is label-based, which means that you have to specify rows and columns based on their row and column labels.

* single label
* list of label
* slice of label
* boolean series
* ; which indicates all the labels

```python
# row 0, all columns
df.loc[0, :]
```

```python
# row 0, 1, 2  [:index] index is inclusive and all columns
df.loc[:2, :]
```

```python
df.loc[[0,1,2], :]
```

```python
# rows 0 - 3 and columns city and state
df.loc[:3, ['City', 'State']]
```

```python
df.loc[:3, 0] # cannot give index to column like this
```

# iloc: iloc is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).
used to select rows and columns (columns by index or integer position). you can pass

* single integer position
* list of integer position
* slice of integer position
# * : which indicates all integer position

```python
# rows position 0 and 1 , column position 0 to 3
df.iloc[:1, :3]
```

```python
df.loc[:1, :]
```

```python
df.loc[:1, ['City', 'Shape Reported']]
```

```python
df.iloc[:1,  ['City', 'Shape Reported']]
```

```python
# rows in position 0 to 2(exclusive) , column 0 to 4

df.iloc[0:2, :4]
```

```python
df.head(1)
```

```python
df.iloc[0:2, 4]
```
