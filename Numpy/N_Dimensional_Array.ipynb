# Random Matrix Generation



*   Uniform Distribution
*   Normal Distribution



````python
import numpy as np
````

# Uniform Distribution

````python
# seed(0)

np.random.seed(5)

np.random.randint(0,10, (3,2))     # Uniform x ∈ [0, 10)
````

# Normal Distribution

````python
np.random.normal(10,2,(3,2))       #  μ = 10   ,   σ = 2
````

````python
np.random.randn(3,2)     # μ = 0   ,   σ = 1
````

# Matrix Multiplication - Linear Algebra / Dot Product in Numpy

````python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8],[9,10],[11,12]])
````

````python
a.shape
````

````python
b.shape
````

````python
np.dot(a,b)      # Dot Matrix Multiplication
````

# Cross Product

````python
c = np.array([3,4,5])
d = np.array([7,8,9])
````

````python
np.cross(c,d)
````
