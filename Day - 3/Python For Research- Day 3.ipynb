```python
# Classwork: WAP to print vowel and consonants from given string
s = 'aeyi'
start = 0
end = len(s)
step = 1

for index in range(start, end, step):
    letter = s[index]

    if letter == 'a' or letter == 'e' or letter == 'i' or letter == 'o' or letter == 'u':
        print("The letter {} is vowel".format(letter))
    else:
        print("The letter {} is consonant".format(letter))
```

## Assignment: WAP to count vowel and consonant in the given string

```python
# Classwork: WAP to print vowel and consonants from given string
s = 'Hi My name is Nirajan'
start = 0
end = len(s)
step = 1

for index in range(start, end, step):
    letter = s[index]

    if letter == 'a' or letter == 'e' or letter == 'i' or letter == 'o' or letter == 'u':
        print("The letter {} is vowel".format(letter))
    elif letter == " ":
        continue       # continue or break. Continue continues the loop while skipping space and break breaks the loop when it finds the space in this code.
    else:
        print("The letter {} is consonant".format(letter))
```
