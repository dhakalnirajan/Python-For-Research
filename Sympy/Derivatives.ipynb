# Higher Order Derivatives
Differentials can take multiple derivatives at once. To take multiple derivatives, pass the variables as many times as you wish to differentiate, or pass a number after the variable.

<img src="https://slideplayer.com/slide/16844340/97/images/5/1.8+Hgher+Order+Derivatives.jpg">

```python
x = symbols('x')
```

```python
fx = x **5 - 3 * x ** 4 + x
```

```python
display(fx)
```

```python
display(diff(fx, x, 5))         # Here, 5 indicates the number of times fx is differentiated with respect to x. The output is the fifth derivative.
```

```python
display(expand(diff(fx, x, 3)))
```

# Partial Derivative

You can also take derivatives with respect to many variables at once. Just pass each derivative in order, using the same syntax as for single variable derivatives. For example, each of the following will compute $$f(x) = e^{xyz}$$

Evaluate:
$$ \frac{\partial^7}{\partial x\partial y^2 z^4} f(x)$$

```python
x, y, z = symbols('x, y, z')
```

```python
fx = exp(x*y*z)
```

```python
display(fx)
```

```python
diff(fx, x, y, y, z, z, z, z)  # Here, y and z are written multiple times. We can also write this way to differentiate fx with respect to y 'n' no. of times and z 'n' no. of times.
```

```python
diff(fx, x, 1, y, 2, z, 4)
```
